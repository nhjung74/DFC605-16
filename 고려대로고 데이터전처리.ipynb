**또한 고려대학교 로고를 라벨링 한다.**

base : 고려대학교 로고
base-label : 고려대학교 로고 라벨
korea : 고려대학교 로고 스라이싱




```
import os, re, glob
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
from matplotlib import pyplot as plt
  
groups_folder_path = '/content/drive/My Drive/Colab Notebooks/'
categories = ["base","base-label","korea"]
 
num_classes = len(categories)

i = 0

for idex, categorie in enumerate(categories):
    label = [0 for i in range(num_classes)]
    label[idex] = 1
    image_dir = groups_folder_path + categorie + '/'
    arrtest = []
    for top, dir, f in os.walk(image_dir):
        dd = 1      
        for filename in f:

            print(image_dir+str(dd)+'.PNG')
            img = Image.open(image_dir+str(dd)+'.PNG')
            data = np.array(img, dtype='float32' )
            dd = dd + 1
            arrtest.append(data)
            #print(arrtest)

    arrtest = np.expand_dims(arrtest, axis=0)
    arrtest.shape
    arrtest /= 255.

    arrtest = np.squeeze(arrtest, axis=0)
    #print(arrtest.shape)
    #print(arrtest)
    np.save(groups_folder_path + categorie + '.npy', arrtest)

imgs = np.load('/content/drive/My Drive/Colab Notebooks/base.npy', allow_pickle=True)

print(imgs.shape)
imgs = np.load('/content/drive/My Drive/Colab Notebooks/base-label.npy', allow_pickle=True)

print(imgs.shape)
imgs = np.load('/content/drive/My Drive/Colab Notebooks/korea.npy', allow_pickle=True)

print(imgs.shape)
```
